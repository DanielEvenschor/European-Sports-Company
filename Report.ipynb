# Introduction

---
# Market Size
Understanding the market size of a country is crucial when looking to expand internationally. 
A larger population often means a large potential customer base, it is also important to consider demographics and competition.

#### Demographics
- Analyzing the demographics of a country is important to understand how the population is segmented, and whether the company's products align with the target audience.

#### Competition
- Examining the competition in a target market is important to determine the potential market share for the company.


---
# Economy
A deep economic assessment of target countries is important to determine if the expansion is reasonable.

#### Purchasing Power
- Examining economic indicators such as GDP or GNI can help understand the purchasing power of the population, we want to determine if our products are affordable in that particular market.

#### Development
- Assessing the development status of a country, as well as its strong industry and service sectors, can help provide a general understanding of the economic conditions.

#### Stability
- Political stability and ease of doing business are also important factors to consider when looking at an expansion.



---
# Other Factors
There are many other factors to consider beyond market size and economic conditions. Some of these factors include:

#### Quality of Life
- The quality of life in a country is important to understand the general living conditions.

#### Climate
- Climate may affect the types of sporting activities that are popular in a given region, as well as the types of clothing that are in demand.

#### Culture
- Culture can affect the types of clothing that people will buy, the types of sports they engage in, and even the colors of the clothing.

### Taxes
- Tax rates can affect the profitability of the expansion and should be taken into consideration

### Ease of doing business
- Some countries might be more complicated to do business in due to differences in process or even corruption

It's important to take all of these factors into consideration when deciding where to expand internationally. Not all problems can be solved by analytics, and some qualitative research may also be necessary to make informed decisions.

Canada offers a large target market of 30 million people living in urban areas. Economic indicators suggest that the population has significant purchasing power left after cost of living. ```python

```python
# GNI per capita     is an     object --> Has to be transformed to FLOAT for calculations -- REMOVE "USD" and ","
allcountrydata["GNI per capita"] = allcountrydata["GNI per capita"].str.replace(" USD", "").str.replace(",", "")
# Convert the remaining values to numeric
allcountrydata["GNI per capita"] = pd.to_numeric(allcountrydata['GNI per capita'], errors = "coerce")
# rename the collumn for clarity that its in USD
allcountrydata = allcountrydata.rename(columns={"GNI per capita": "GNI ($ per capita)"})
# "World Share" --> remove " %" then make float - add "in %" to collumn name
allcountrydata["World Share"] = allcountrydata["World Share"].str.replace(" %", "")
# Convert the data in "World Share" to a float
allcountrydata["World Share"] = allcountrydata["World Share"].astype(float)
# Rename "World Share" to "World Share in %"
allcountrydata = allcountrydata.rename(columns={"World Share": "World Share in %"})
# Same for "Yearly Change" remove % and add to collumn title
allcountrydata["Yearly Change"] = allcountrydata["Yearly Change"].str.replace(" %", "")
# Convert the data in "Yearly Change" to a float
allcountrydata["Yearly Change"] = allcountrydata["Yearly Change"].astype(float)
# Rename "Yearly Change" to "Yearly Change in %"
allcountrydata = allcountrydata.rename(columns={"Yearly Change": "Yearly Change in %"})
# "Industry", "Service" need to be floats, Values look like this : 0,12 - either run astype or remove 0, and make it % in title
allcountrydata["Industry"] = allcountrydata["Industry"].str.replace(",", ".")
allcountrydata["Industry"] = pd.to_numeric(allcountrydata['Industry'], errors = "coerce")*100
allcountrydata = allcountrydata.rename(columns={"Industry": "Industry in %"})
allcountrydata["Service"] = allcountrydata["Service"].str.replace(",", ".")
allcountrydata["Service"] = pd.to_numeric(allcountrydata['Service'], errors = "coerce")*100
allcountrydata = allcountrydata.rename(columns={"Service": "Service in %"})
# "Urban Pop %" remove % and make it a float
allcountrydata["Urban Pop %"] = allcountrydata["Urban Pop %"].str.replace(" %", "")
allcountrydata["Urban Pop %"] = pd.to_numeric(allcountrydata['Urban Pop %'], errors = "coerce")
# "Age above 65 Years", "Age 15 to 64 Years", "Age 0 to 14 Years" remove % and add to title
allcountrydata["Age above 65 Years"] = allcountrydata["Age above 65 Years"].str.replace("%", "").astype(float)
allcountrydata["Age 15 to 64 Years"] = allcountrydata["Age 15 to 64 Years"].str.replace("%", "").astype(float)
allcountrydata["Age 0 to 14 Years"] = allcountrydata["Age 0 to 14 Years"].str.replace("%", "").astype(float)
allcountrydata = allcountrydata.rename(columns={"Age above 65 Years": "Age above 65 Years in %"})
allcountrydata = allcountrydata.rename(columns={"Age 15 to 64 Years": "Age 15 to 64 Years in %"})
allcountrydata = allcountrydata.rename(columns={"Age 0 to 14 Years": "Age 0 to 14 Years"})
# "Med. Age" to numeric coerce erros
allcountrydata["Med. Age"] = pd.to_numeric(allcountrydata['Med. Age'], errors = "coerce")
```

```python
# Previous graphs gave me some insights but its very hard to exclude a certain country based on this information
# Decided to create my own parameter: Purchasing power - Cost of living
# Logically the higher the number, the more money people have left for buying sporting goods and other things
# Find the top 5 countries for this metric 
# Furthermore i want to have a deeper look at demographic, since most our customers are in the age range 15 to 64 this is the most important group
# I still want to see 0 to 14 years old because it can give me an idea how the age structure might change 

# Create new collumn, subtracting existing ones 
selection["Purchasing Power - Cost of Living"] = selection["Purchasing Power Index"] - selection["Cost of Living Index"]
# Filter top 5 for that collumn 
filter_2 = selection[["Country Name", "Purchasing Power - Cost of Living", "Age 0 to 14 Years",
           "Age 15 to 64 Years in %"]].sort_values(by='Purchasing Power - Cost of Living',ascending=False)[:5]

# Plot the data using seaborn 
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(filter_2['Country Name'])) # arrange according to the filter 
bar_width = 0.35

# ignore the variable names, i changed the color after
# Here i split the two bars and make them different colors 
blue_bars = ax.bar(x - bar_width/2, filter_2['Age 0 to 14 Years'], bar_width, color='brown', label='Age 0 to 14 Years')
red_bars = ax.bar(x + bar_width/2, filter_2['Age 15 to 64 Years in %'], bar_width, color='black', label='Age 15 to 64 Years in %')

# Adding labels
ax.set_xticks(x)
ax.set_xticklabels(filter_2['Country Name'])
ax.set_xlabel('Country Names', fontsize=12)
ax.set_ylabel('% of Population', fontsize=12)
ax.set_title("Top 5 Countries for 'Purchasing Power - Cost of Living'")
ax.legend(loc='upper left')

# Add values on top of bars
for blue_bar, red_bar in zip(blue_bars, red_bars):
    height_blue = blue_bar.get_height()
    height_red = red_bar.get_height()
    ax.text(blue_bar.get_x() + blue_bar.get_width()/2, height_blue + 1, f"{height_blue:.1f}%", ha="center", va="bottom")
    ax.text(red_bar.get_x() + red_bar.get_width()/2, height_red + 1, f"{height_red:.1f}%", ha="center", va="bottom")

# Find the highest value
max_value1 = filter_2['Age 0 to 14 Years'].max()
max_value2 = filter_2['Age 15 to 64 Years in %'].max()
coordinate1 = 1.82,max_value1+8 # unfortunately this is static, i couldnt find a way to make the icon automatically find the right coordinates
coordinate2 = 3.15, max_value2+8

ax.scatter(*coordinate1, marker='v', c='green', ec='green', lw=5, s=200) # customize the icon 
ax.scatter(*coordinate2, marker='v', c='green', ec='green', lw=5, s=200)

plt.show()
# This graph reduces us to 5 countries: USA, Australia, South Africa, United Arab Emirates and Canada
# I highlight UAE and South Africa because they have the biggest % of the population in our target demographic
# South africa with a very high % of very young people, wich will naturally get older and move to our target demographic
```
